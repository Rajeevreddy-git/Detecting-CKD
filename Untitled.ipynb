```python
import pandas as pd
```

```python
df=pd.read_csv("kidy.csv")
```

```python
df
```

```python
df_doubled = pd.concat([df, df], ignore_index=True)  # Simply double it

df_doubled.to_csv("doubled_dataset.csv", index=False)
```

```python
df_doubled
```

```python
cols_drop = ['id', 'ba', 'pcc', 'pc', 'su', 'cad', 'pe', 'ane', 'appet', 'dm']  # You can customize based on missing values
df_clean = df_doubled.drop(columns=cols_drop)
```

```python
df_clean
```

```python
df_clean.isnull()
```

```python
import numpy as np
```

```python
df_clean.replace(to_replace=['\t\?', '\?', '--', ' '], value=np.nan, regex=True, inplace=True)
```

```python
df_clean.isnull().sum()
```

```python
# First, let's clean the columns that should have numeric values.
num_cols = ['age', 'bp', 'sg', 'al', 'rbc', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']

# Convert to numeric and fill NaN with median
for col in num_cols:
    # Convert all values in column to numeric, invalid parsing will be set to NaN
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
    
    # Now, fill missing values (NaN) with the median of the column
    df_clean[col].fillna(df_clean[col].median(), inplace=True)
```

```python
df_clean.isnull().sum()
```

```python
df_clean.drop(columns=['rbc'], inplace=True)
```

```python
df_clean.isnull().sum()
```

```python
df_clean
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
le = LabelEncoder()

# Apply label encoding to the 'classification' and 'htn' columns
df_clean['classification'] = le.fit_transform(df_clean['classification'])
df_clean['htn'] = le.fit_transform(df_clean['htn'])
```

```python
df_clean
```

```python
df_clean.shape
```

```python
x=df_clean.iloc[:,0:-1]
y=df_clean.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y ,random_state=10 )
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

clf_gini=DecisionTreeClassifier(criterion='gini',random_state=0)
clf_gini.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
p=clf_gini.predict(x_test)
```

```python
print(accuracy_score(p,y_test))
```

    # For numeric columns, enter the values directly as integers or floats

    bp = float(input("Enter BP (Blood Pressure): "))  # BP (e.g., 80)

    al = float(input("Enter AL (Albumin): "))  # AL (e.g., 1.0)

    bu = float(input("Enter BU (Blood Urea): "))  # BU (e.g., 36.0)
    sc = float(input("Enter SC (Serum Creatinine): "))  # SC (e.g., 1.2)
    sod = float(input("Enter SOD (Sodium): "))  # SOD (e.g., 138.0)

    htn = int(input("Enter HTN (Hypertension - 0 or 1): "))  # HTN (0 or 1)

    # Return the values as a list (matching the model's expected input format)
    return [[age, bp, sg, al, bgr, bu, sc, sod, pot, hemo, pcv, wc, rc, htn]]
